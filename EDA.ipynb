```python
import seaborn as sns
correlation=df.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)

#Resulting inference :
#importing seaborn library for data visualization
#Building correlation matrix for to see relationship between any 2 attributes in the dataset
```

```python
sns.distplot(df['COVID-19 deaths per 100,000 population in 2020'])


#Resulting inference :
#The distplot represents the univariate distribution of data i.e. 
#data distribution of a variable against the density distribution.
#From the graph we can see that if covid-19 deaths per 1 lakh population in 2020 is greater than 0 but less than 50
#then their distribution in data is between 0.005 and 0.035
#similarly we can conclude for other points 
```

```python
sns.catplot(x='COVID-19 deaths per 100,000 population in 2020', kind='box',data=df)


#Resulting inference :
#Median value of covid-19 deaths per 1 lakh population in 2020 is between 0 and 25
#Minimum value of x axis variable is 0 and 25th percentile is near to zero
#Maximum value of x axis variable is between 100 and 125 
#75th percentile of x-axis is less than 50
#Outliers are greater than 100

```

```python
sns.relplot(x='Population 2020',y='COVID-19 deaths per 100,000 population in 2020',data=df)

#Resulting inference:
#Relation between population of 2020 and covid-19 per 1 lakh population in 2020 is shown
#We can see that countries with population between 10^9 to 0.2*10^9 have covid-19 deaths per 1 lakh population
#between 0 to 175 
#2 countries with population around 1.4*10^9 report covid-19 deaths per 1 lakh between 0 to 25 
```
